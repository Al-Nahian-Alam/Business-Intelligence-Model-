#2.Find the store_size-wise monthly total sales price of all stores joining the fact table and respective dimension table and visualize it to a bar chart

```python
add_time_with_fact = pd.merge(pd.merge(fact_table, time_dim, on='time_key'),
store_dim, on='store_key')
add_time_with_fact
```

```python
store_size_wise_monthly_total_sales = add_time_with_fact.groupby(['store_size','month']).aggregate({'total_price':'sum'}).reset_index()

```

```python
store_size_wise_monthly_total_sales
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12, 6))
plt.title('Store_size-wise Monthly total sales price')
sns.barplot(x='month', y='total_price', hue='store_size', data=store_size_wise_monthly_total_sales)
```

#4. What are the average sales price of items per store quarterly? Show the data in a line chart

```python
add_time_with_fact = pd.merge(pd.merge(fact_table, time_dim, on='time_key'),
store_dim, on='store_key')
add_time_with_fact
```

```python
store_wise_quarterly_average_sales = add_time_with_fact.groupby(['store_key','quarter']).aggregate({'total_price':'mean'}).reset_index()
```

```python
store_wise_quarterly_average_sales
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.figure(figsize=(12, 8))
plt.title('Average sales price of items per store quarterly')
sns.lineplot(x='quarter', y= 'total_price',hue='store_key', data=store_wise_quarterly_average_sales)
```
